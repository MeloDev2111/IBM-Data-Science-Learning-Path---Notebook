# Extracting Stock Data Using a Python Lib

## Business Understanding

A company stock, also known as a share or equity, is a portion of ownership in a company. When you buy a company's stock, you become a shareholder and are entitled to a portion of the company's profits.

An investor can buy a stock and sell it later. If the stock increase in value, the investor can sell the stock for a profit. If it decreases, the investor will incur in a loss. Determining the stock price is complex; it depends on the number of outstanding shares, the size of the company's future profits, and much more. People trade stocks throughout the day.

The **stock ticker** is a report of the price of a certain stock, updated continuously throughout the trading session by the various stock market exchanges.

### Extract Apple Stock Data Using yfinance

```python
# Instance Ticker module for Apple(AAPL)
apple = yf.Ticker("AAPL")
```

```python
# Stock info
apple.info
```

Extracting Share Price History

```python
# Extracting Share Price
## Obtaining the share price of the stock with a parameter of period = max
apple_share_price_data = apple.history(period="max")

## Columns pre reset of index - Index is the date
apple_share_price_data.columns
## Resetting index
apple_share_price_data.reset_index(inplace=True)
## Columns post reset of index
apple_share_price_data.columns

## Displaying the first 5 rows of the data
apple_share_price_data.head()
```

```python
## Plotting Open vs Date
apple_share_price_data.plot(x="Date", y="Open")
```

### Extract AMD Stock Data

```python
# Instance Ticker
amd = yf.Ticker("AMD")
amd.info
```

```python
# Showing some basic info
print(f"Country: {amd.info['country']}")
print(f"Sector: {amd.info['sector']}")
```

```python
# Extracting Share Price
amd_share_price_data = amd.history(period="max")
amd_share_price_data.head()
```

```python
## Find volume traded on the first day
### Sorting the data in ascending order - The date is the index
amd_share_price_data.sort_index(ascending=True, inplace=True)
print(amd_share_price_data.head())
print("Volume trades on the first day: " + amd_share_price_data["Volume"].iloc[0].astype(str))
```
