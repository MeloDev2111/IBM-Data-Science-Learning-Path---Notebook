### Grouping

```python
# define variables to group
df_group_one = df[['drive-wheels', 'body-style', 'price']]
df_group_one
```

```python
# grouping results
# calculate avg price for each of the different categories of drive-wheels
df_grouped = df_group_one.groupby(['drive-wheels'], as_index=False).agg({'price': 'mean'})
df_grouped
```

```python
# can also group by multiple variables
df_group_multi_var = df[['drive-wheels','body-style', 'price']]
df_grouped_multi = df_group_multi_var.groupby(['drive-wheels', 'body-style'], as_index=False).mean()
df_grouped_multi
```

```python
# Use pivot for visualize grouped data
grouped_pivot = df_grouped_multi.pivot(index='drive-wheels', columns='body-style')
grouped_pivot.fillna(0, inplace=True)
grouped_pivot
```

```python
# Visualizing the relationship between Drive-wheels, Body Style vs Price Using a HeatMap
plt.pcolor(grouped_pivot, cmap='RdBu') #Red Blue
plt.colorbar()
plt.show()
```

```python
# add labels
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu') #Red Blue

# label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

# move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False) #x # rows
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False) #y # cols

# insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

# rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

### Correlation and Causation
**Correlation**: a measure of the extent of interdependence between variables
**Causation**: the relationship between cause and effect between two variables

Correlation does not imply causation. Determining correlation is much simpler than determining causation as causation may require independent experimentation

### Methods for measure correlation
- **Pearson Correlation** measure the line dependence between two variables, the resulting coefficient is a value between -1 and 1 inclusive. **default method used in the function corr**
    - **1**: Perfect positive linear correlation
    - **0**: no linear correlation, the two vars most likely do not affect each other
    - **-1**: Perfect negative correlation.

- **P-value** is the probability value that the correlation between these two variables is statistically significant. using a significance level of 0.05 normally.
    - **<0.001**: we say there is **strong** evidence that the correlation is significant
    - **<0.05**: there is **moderate** evidence that the correlation is significant
    - **<0.1**: there is **weak** evidence
    - **>0.1**; there is **no evidence**

- **Chi-Square test** for categorical variables, research more about this. chi-square value, degrees of freedom and significance level (normally 0.05)

**We can obtain this measures using the "stats" module in the "scipy" lib**

```python
from scipy import stats
# Wheel-base vs price
pearson_coef, p_value = stats.pearsonr(df['wheel-base'], df['price'])
print("The Pearson correlation coefficient is", pearson_coef, "with a P-value of P = ", p_value)
# Since the p-value is <0.001,  the correlation between both variable is statistically significant,
# although the linear relationship isn't extremely strong (0.59) not too close to 1
```
