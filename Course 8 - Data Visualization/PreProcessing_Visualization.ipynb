# Data Cleaning

```python
# Remove a few unnecessary columns
# in pandas axis=0 represents rows (default) and axis=1 represents columns.
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df_can.head(2)
```

```python
# Rename Columns
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_can.columns
```

```python
# Add a "Total" column that sums up to the total immigrants by country
years = list(map(int, range(1980, 2014))) # from 1980 to 2014
df_can['Total'] = df_can[years].sum(axis=1)
df_can['Total']
```

```python
# Check null objects
df_can.isnull().sum()
```

```python
df_can.describe()
```

```python
# 1. the full row data (all columns)
df_can.loc['Japan'] # form 1
df_can.iloc[87] # form 2
df_can[df_can.index == 'Japan'] # form 3
```

# Data formatting

```python
# let's convert the column names into strings: '1980' to '2013'.
print("columns before formatting: ",df_can.columns.to_list()) 
df_can.columns = list(map(str, df_can.columns))
print("columns after formatting: ",df_can.columns.to_list())
```
